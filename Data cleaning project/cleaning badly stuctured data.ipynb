```python
# import the required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set()
```

```python
#load the dataset
df = pd.read_excel('Badly-Structured-Sales-Data-2.xlsx')
```

```python
df.head()
```

```python
#Rename the columns
df = df.rename(columns={'Unnamed: 2': 'First Class', 'Unnamed: 3': 'First Class',
                       'Unnamed: 5': 'Same Day', 'Unnamed: 6': 'Same Day',
                       'Unnamed: 8': 'Second Class', 'Unnamed: 9': 'Second Class',
                       'Unnamed: 11': 'Standard Class', 'Unnamed: 12': 'Standard Class'})
df.head()
```

```python
#Transpose columns to rows and rows to columns
df_t = df.transpose()
df_t
```

```python
#Make the first row the header
df_t.columns = df_t.iloc[0]
df_t = df_t.iloc[1:].reset_index(drop=False)
df_t.head()
```

```python
#Use pd.melt to shrink the columns
#This converts the wide dataset to a long one
df_melt = pd.melt(df_t, id_vars=['index', 'Segment'], var_name='Order Date', value_name='Sales')
df_melt
```

```python
#Drop the nan values
df_melt = df_melt.dropna()
df_melt
```

```python
#Rename the index column to Ship Mode
df_melt = df_melt.rename(columns={'index':'Ship Mode'})
df_melt.head()
```

```python
#Reset index making the drop arguement True so that pandas can drop the old index
df_melt = df_melt.reset_index(drop=True)
df_melt
```

## New Task

Lets see in which part of the year and month does the business receive more patronage

```python
#Lets make a copy of the original dataframe and save it in a variable called calendar df
calendar_df = df_melt.copy()
calendar_df.head()
```

```python
#Check if the order date column is a datetime dtype
calendar_df.info()
```

```python
#Convert the order dates column to a datetime dtype
calendar_df['Order Date'] = pd.to_datetime(calendar_df['Order Date'])
```

```python
#Check for implemented changes
calendar_df.info()
```

```python
#Lets split the order dates column into day, month and year
import calendar
calendar_df['Day'] = calendar_df['Order Date'].dt.day
calendar_df['Month'] = calendar_df['Order Date'].dt.month
calendar_df['Year'] = calendar_df['Order Date'].dt.year

calendar_df.head()
```

```python
#Lets convert the months column to month names instead of numbers
#First we import the calendar library
#Next we use the apply lambda function to apply the library to that column
import calendar

calendar_df['Month'] = calendar_df['Month'].apply(lambda x: calendar.month_name[x])

calendar_df.head()
```

```python
#Which year had more customer patronage
calendar_df['Year'].value_counts().plot(kind='bar', figsize=(14,6))
```

```python
#Check the unique values in the Month column using .unique()
calendar_df.Month.unique()
```

```python
#Lets plot a bar graph that shows which months has the highest sales activity
calendar_df.Month.value_counts().plot(kind='bar', figsize=(14,6))
```
